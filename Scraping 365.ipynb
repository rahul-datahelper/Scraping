# currency convertor

```python
date= input("Enter the date (in the format 'yyyy-mm-dd' or 'latest'): ")
base= input("convert from (currency): ")
curr= input("convert to (currency): ")
quan= float(input('How much {} do you want to convert: '.format(base)))

url=b_url + "/" + date + "?base="+ base + "&symbols=" + curr
response= requests.get(url)

if (response.ok is False):
    print("error ".format(response.status_code))
    print(response.json()['error'])
    
else:
    data= response.json()
    rate=data['rates'][curr]
    
    result=quan*rate
    print("\n {1} is equal to {2} {3} , based upon exchange rates on {4}".format(quan,base,result,curr,data['date']))
```

# Pagination

# extracting results from multiple pages

```python
results=[]
```

```python
for i in range(5):
    r=requests.get(base_url,params= {"page": i+1})
    
    if len(r.json())==0:
        break
        
    else:
        results.extend(r.json())
```
